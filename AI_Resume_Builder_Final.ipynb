```python
from fpdf import FPDF

print("===== AI Resume Builder =====")

name = input("Enter your name: ")
education = input("Enter your education: ")
skills = input("Enter your skills: ")
projects = input("Enter your projects: ")
experience = input("Enter your experience: ")

# Generate Resume Content
resume_text = f"""
{name}

Career Objective:
Motivated and enthusiastic Information Technology student seeking opportunities to apply technical and communication skills in a dynamic organization.

Education:
{education}

Skills:
{skills}

Projects:
{projects}

Experience:
{experience}

Declaration:
I hereby declare that the above information is true to the best of my knowledge.
"""

print("\nGenerated Resume:\n")
print(resume_text)

# Create PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8, resume_text)

pdf.output("Vadla_Mamatha_Resume.pdf")

print("\n✅ Resume PDF Created Successfully!")
```

```python
from fpdf import FPDF

print("===== AI Resume & Portfolio Builder =====")

name = input("Enter your name: ")
education = input("Enter your education: ")
skills = input("Enter your skills (comma separated): ")
projects = input("Enter your projects: ")
experience = input("Enter your experience: ")

# Convert skills into bullet format
skill_list = skills.split(",")

formatted_skills = ""
for skill in skill_list:
    formatted_skills += f"- {skill.strip()}\n"

resume_text = f"""
{name.upper()}

CAREER OBJECTIVE
Motivated and dedicated Information Technology student with strong problem-solving abilities and communication skills. Seeking an opportunity to contribute technical knowledge and grow professionally.

EDUCATION
{education}

TECHNICAL SKILLS
{formatted_skills}

PROJECTS
{projects}

EXPERIENCE
{experience}

DECLARATION
I hereby declare that the information provided above is true to the best of my knowledge.
"""

print("\nGenerated Resume:\n")
print(resume_text)

# Create PDF
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.multi_cell(0, 8, resume_text)

file_name = name.replace(" ", "_") + "_Resume.pdf"
pdf.output(file_name)

print(f"\n✅ Resume PDF '{file_name}' Created Successfully!")
```
